```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
# model layers are arranged one after anothr
from tensorflow.keras.layers import Dense
#dense means fully connected layers
from tensorflow.keras.optimizers import Adam
#it update weights to reduce loss
```

```python
# Spam Message Classification
data = {
    "message_length": [20, 150, 35, 200, 50, 250, 40, 300],
    "links_count": [0, 5, 1, 8, 1, 10, 0, 12],
    "special_chars": [1, 15, 2, 20, 3, 25, 1, 30],
    "spam": [0, 1, 0, 1, 0, 1, 0, 1]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
X = df.drop("spam",axis=1)
Y= df["spam"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()
```

```python
model.add(Dense(16,input_dim=3,activation ='relu'))
```

```python
model.add(Dense(8, activation='relu'))
```

```python
model.add(Dense(1, activation='sigmoid')) #used for binary classification
```

```python
model.compile(
    loss='binary_crossentropy',
    optimizer=Adam(learning_rate=0.01),
    metrics=['accuracy']
)
```

```python
model.fit(X_train, Y_train, epochs=20, batch_size=4)
```

```python
loss,accuracy = model.evaluate(X_test,Y_test)
print(accuracy)
```

```python
new_student = np.array([[6, 85, 70]]) # study_hours, attendance, previous_marks
new_student = scaler.transform(new_student)
```

```python
prediction = model.predict(new_student)
```

```python
if prediction >0.5:
    print("Spam Message")
else:
    print("Non Spam Message")
```
