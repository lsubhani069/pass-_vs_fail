```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans # elbow method to find k
```

```python
data = {
    "TasksCompleted": [5,6,7,8,9,15,16,18,20,22,30,32,35,38],
    "WorkHours": [4,5,5.5,6,6.5,7,7.5,8,8.5,9,9.5,10,10.5,11],
    "Rating": [2,2.5,3,3.2,3.5,3.8,4,4.2,4.5,4.7,4.8,5,5,5]
}
```

```python
df = pd.DataFrame(data)
```

```python
wcss=[] #within cluster sum of square
for k in range(1,7):
    km = KMeans(n_clusters=k,random_state=42) #model = logisticRegression()
    #km=KMeans()
    km.fit(df)
    wcss.append(km.inertia_)
plt.plot(range(1,7),wcss,marker='x')
plt.title("Elbow")
plt.xlabel("Number of Clusters")
plt.ylabel("Wcss")
plt.show()
```

```python
model = KMeans(n_clusters=3,random_state=0)
df['Cluster']=model.fit_predict(df)
```

```python
centroids = model.cluster_centers_
```

```python
print(pd.DataFrame(centroids, columns=df.columns[:-1])) #centroid data
```

```python
plt.figure()
```

```python
plt.scatter(df['TasksCompleted'], df['WorkHours'], c=df['Cluster'])

plt.scatter(centroids[:,0], centroids[:,1], marker='o', s=300)   #all row,column 0,all rows , column 1
plt.title("K-Means Clustering")
plt.ylabel("Work Hours")
plt.xlabel("Tasks Completed")
```
